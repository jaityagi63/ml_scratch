```python
import numpy as np

class GDregressor:
    def __init__(self, m, lr=0.001, epoch=100):
        self.lr = lr
        self.epoch = epoch
        self.m = 12.50110143               # slope must be passed explicitly
        self.b = float(input('Enter value of b: '))  # initial bias from user input

    def fit(self, x, y):

        for i in range(self.epoch):
            # Compute average gradient w.r.t b
            loss_slope = -2 * np.sum(y - self.m * x.ravel() - self.b)
            step_size = loss_slope * self.lr
            self.b = self.b - step_size
            y_pred = self.m * x + self.b
            plt.plot(x,y_pred,color='blue')
            plt.scatter(x,y,color='red')
            plt.plot(x,reg.predict(x),color='red')

            print(f'epoch {i+1} : b = {self.b:.5f}')
            print(f"loss_slope: {loss_slope:.5f}, step_size: {step_size:.5f}, current b: {self.b:.5f}")

        return self.b

         # Use self.m
```
