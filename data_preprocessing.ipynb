#normalization


```python
#when data is numerical
import numpy as np

data = np.array([1,2,3,4,5,6,7,8,11,23,45,100,2000])

# Now there are many technique but the one I used are:
# 1.min-max scaling ( mainly used when data is not guassian distributed and this is good when there is range )
# 2.Z-Score Normalization( this is widely used in ml and it is better then min-max normalisation)
# 3.Robust Scaling (never used it so don't know but it has something to do with quartile go figure it out yourself )

def min_max_scaling(data):
  min = np.min(data)
  max = np.max(data)
  return (data - min) / (max - min)

def z_score_normalization(data):
  mean = np.mean(data)
  std = np.std(data)
  return (data - mean) / std
```

```python
print(f'orignak data: {data}\n')
print(f'min-max scaling: {min_max_scaling(data)}\n')
print(f'z_score_scaling: {z_score_normalization(data)}')
```

```python
import plotly.express as px
import pandas as pd

df = pd.DataFrame({
    'Index': range(len(data)),
    'Original': data,
    'Min-Max': min_max_scaling(data),
    'Z-Score': z_score_normalization(data)
})

df_melted = df.melt(id_vars=['Index'], var_name='Scaling Method', value_name='Value')

fig = px.line(
    df_melted,
    x='Index',
    y='Value',
    color='Scaling Method',
    title='Comparison of Data Scaling Techniques'
)

fig.update_layout(
    hovermode='x unified',
    template='plotly_white'
)

fig.show()
```
