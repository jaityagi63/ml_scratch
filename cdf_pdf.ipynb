```python
import pandas as pd
import numpy as np
import random


L = []
for i in range(1000000):
    a = random.randint(1,6)
    b = random.randint(1,6)
    L.append( a + b )

data = pd.Series(L)
print(data.value_counts(),end=" ")

percenatge = (data.value_counts()/data.value_counts().sum()).sort_index()
percenatge.plot(kind='bar')
```

```python
np.cumsum(percenatge).plot(kind='bar')
```

```python
import matplotlib.pyplot as plt
from numpy.random import normal

data = normal(loc=50, scale=5, size=1000)
plt.hist(data,bins=10)

sample_mean = data.mean()
sample_std = data.std()
```

```python
from scipy.stats import norm

dist = norm(sample_mean,sample_std)
values = np.linspace(data.min(),data.max(),100)
prob = [ dist.pdf(value) for value in values]

plt.hist(data,bins=10,density=True)
plt.plot(values,prob)
```

```python
sample1 = normal(loc=20,scale=5,size=300)
sample2 = normal(loc=40,scale=5,size=700)
sample = np.hstack((sample1,sample2))
plt.hist(sample,bins=50)
```

```python
from sklearn.neighbors import KernelDensity

kde = KernelDensity(bandwidth=2,kernel='gaussian')
sample = sample.reshape(len(sample),1)
kde.fit(sample)

values = np.linspace(sample.min(),sample.max(),100)
values = values.reshape((len(values),1))
probalit = kde.score_samples(values)
prob = np.exp(probalit)

plt.hist(sample,bins=40,density=True)
plt.plot(values[:],prob)
plt.show()
```
